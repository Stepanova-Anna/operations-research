## 2. Постановка задачи

Обозначим:

- $x_1$ — количество участков ямочного ремонта;
- $x_2$ — количество участков замены бордюров.

Целевая функция:

$$
\max z = 8x_1 + 7x_2
$$

Ограничения:

$$
3x_1 + 2x_2 \le 24 \quad \text{(бригада)}
$$

$$
2x_1 + 3x_2 \le 30 \quad \text{(техника)}
$$

$$
x_1 \ge 0, \quad x_2 \ge 0
$$

### Числа из условия

- полезность ямочного ремонта: `8`
- полезность замены бордюров: `7`
- бригада: `3` и `2`, лимит `24`
- техника: `2` и `3`, лимит `30`

## 4. Задание 2. Проверка через `linprog`

Здесь нужно вручную заполнить массивы для решателя.

Подсказка по порядку работы:

1. Сначала перепишите коэффициенты в `c`, `A_ub`, `b_ub`.
2. Затем проверьте, что вы правильно поменяли знак у целевой функции.
3. Только после этого запускайте решатель.

```python
import numpy as np
from scipy.optimize import linprog


def describe_result(result) -> None:
    """Печатает найденный план в удобном для отчёта виде.

    Args:
        result: Объект результата от `linprog`.
    """
    print('success =', result.success)
    print('message =', result.message)
    if result.success:
        print('x =', result.x)
        print('f_min =', result.fun)
        print('z_max =', -result.fun)


# Коэффициенты целевой функции для ЗАДАЧИ МИНИМИЗАЦИИ (-z)
c = np.array([-8, -7], dtype=float)

# Матрица ограничений A_ub @ x <= b_ub
A_ub = np.array([
    [3, 2],
    [2, 3],
], dtype=float)

# Правые части ограничений
b_ub = np.array([24, 30], dtype=float)

bounds = [(0, None), (0, None)]

if np.isnan(c).any() or np.isnan(A_ub).any() or np.isnan(b_ub).any():
    print('Сначала заполните c, A_ub и b_ub числами из постановки.')
else:
    result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=bounds, method='highs')
    describe_result(result)
```
